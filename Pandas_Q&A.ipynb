```python
import pandas as pd
```

```python
# Using iloc, loc, and ix

ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.head()
```

```python
# loc is for selecting and filtering rows and columns by label

ufo.loc[0, :]
```

```python
ufo.loc[[0, 1, 2], :]
```

```python
ufo.loc[0:2, :] # loc is inclusive on both sides of the colon
```

```python
# loc column selection

ufo.loc[:, ['City', 'State']]
```

```python
ufo.loc[0:2, 'City' : 'State'] # 0:2 is inclusive on both sides
```

```python
ufo.head(3).drop('Time', axis=1)
```

```python
# Use Boolean values

ufo[ufo.City=='Oakland']
```

```python
ufo.loc[ufo.City=='Oakland', :]
```

```python
# iloc is used for filtering rows and columns using integer position

ufo.iloc[:, [0,3]]
```

```python
ufo.iloc[:, 0:4] # iloc is exclusive of the 4

# similar to list(range(0,4))
```

```python
ufo.iloc[0:3, :]
```
